### Non Negative Axiom

```python
sample_space={1,2,3,4,5,6}
probabilities={outcomes:1/6 for outcomes in sample_space}

for outcomes,prob in probabilities.items():
    print(f"p({outcomes})={prob} (>=0)")
print("--"*20)
```

## join probability

```python
#rolling a dice
sample_space=[1,2,3,4,5,6]

A=[2,4,6] # even numbers
B=[4,5,6] #grter than 3

intersection=[i for i in sample_space if i in A and i in B ]
P_A_intersection_B=len(intersection)/len(sample_space)

print("Event A:",A)
print("Event B:",B)
print("A intersection B:",intersection)
print("P(A intersection B):",P_A_intersection_B)
```

## Discrete Probability Distrubution (PMF,CDF)

```python
scores=[1,2,2,3,3,3,4,4,5]

unique_scores=sorted(set(scores))
print("Unique scores:",unique_scores)

#caluclate PMF (probability of each unique scores)
pmf=[scores.count(x)/len(scores) for x in unique_scores]

#Caluclate CDF(cumultive sum of PMF values)
cdf=np.cumsum(pmf)
print()
print("Score | pmf | cdf")
for  x,p,f in zip(unique_scores,pmf,cdf):
    print(f"{x}      | {p:3f} |{f:3f}")

# Plot pmf using bar chat
plt.bar(unique_scores,pmf,color="skyblue",edgecolor="black",label="PMF")

#plot cdf using step
plt.step(unique_scores,cdf,where="post",color="red",linewidth=2,label="CDF")

plt.title("Discrete Probability Distrubution (PMF,CDF)")
plt.xlabel("Score")
plt.ylabel("Probability")
plt.ylim(0,1.05)
plt.legend()
plt.grid(True)
plt.show()
```

## Multinomial Distribution

```python
import matplotlib.pyplot as plt
from math import factorial
from functools import reduce
import operator
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
n=10
count=[2,3,1,0,3,1] #1 to 6
p=[1/6]*6

#formula
denominator=reduce(operator.mul,(factorial(x) for x in count),1)
multinomial_coefficient=factorial(n)/denominator
print(multinomial_coefficient)

sequence_prob=reduce(operator.mul,(p[i]**count[i] for i in range(6)),1)
print(sequence_prob)

probability=multinomial_coefficient * sequence_prob
print(probability)
print(f"{1/probability:.0f}")

#plot
faces=[1,2,3,4,5,6]
plt.figure(figsize=(8,5))
sns.barplot(x=faces,y=count,palette="pastel")
plt.title("Dice roll outcomes count(10 rolls)")
plt.xlabel("Die Faces")
plt.ylabel("NO.of Occurrences")
plt.ylim(0,max(count)+1)
plt.text(2.5,max(count)+0.5,f"probability={probability:.6f}",fontsize=12,color="red")
plt.show
```
