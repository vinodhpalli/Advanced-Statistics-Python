## P Value

```python
import numpy as np
from scipy import stats

data=[52,55,60,58,57,54,59,53]
pop_mu=50

# t test
t_stat,p_value=stats.ttest_1samp(data,pop_mu)

print(t_stat)
print(p_value)
```

## CDF

```python
from scipy.stats import norm

#mean=5,std=1
p_less_than_3=norm.cdf(3,5,1)
p_less_than_4=norm.cdf(4,5,1)
p_less_than_5=norm.cdf(5,5,1)
p_less_than_6=norm.cdf(6,5,1)
p_less_than_7=norm.cdf(7,5,1)
p_less_than_8=norm.cdf(8,5,1)
p_less_than_9=norm.cdf(9,5,1)

print(p_less_than_3) 
print(p_less_than_4) #15.87% of customer waits less than 4 minutes
print(p_less_than_5) 
print(p_less_than_6) #84.13% of customer waits less than 6 minutes
print(p_less_than_7) 
print(p_less_than_8) 
print(p_less_than_9) 
```
