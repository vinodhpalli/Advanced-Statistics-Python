## Probability

### Tossing a coin

```python
outcomes=["Head","Tail"]
favourable_event=["Head"] #Gtting Head

#probability ----> favourable_event/total

p_head=len(favourable_event)/len(outcomes)
print("ptobability of getting head:",p_head)
```

```python
#RE,Trail,ss,Event
#Random Experiment :roling a dice

import random

dice_roll=random.randint(1,6)
print("Dice roll outcome:",dice_roll)

#Sample Space
sample_space=[1,2,3,4,5,6]
print("Sample Space:",sample_space)

#Event :getting even number
event_even=[2,4,6]
p_even=len(event_even)/len(sample_space)
print("Probability of even number:",p_even)
```

## Event

### Simple Event

```python
#tossing coin once
outcomes=["Head","Tail"]
#simple_event:Gtting Head

#probability ----> favourable_event/total

p_head=1/len(outcomes)
print("ptobability of Simple event(head):",p_head)
```

### Compund event

```python
import itertools #generate all posible combinations

coins=["Head","Tail"]

sample_space=list(itertools.product(coins,repeat=2))
print("sample_space:",sample_space)

#copund Event:Getting one head atlest

compound_event=[i for i in sample_space if "Head" in i]
print("Getting atleast one Head:",compound_event)

#probability of compound event
p_compound=len(compound_event)/len(sample_space)
print("probability of compound event:",p_compound)
```

### Certain Event

```python
coins=["Head","Tail"]
certain_event=["Head","Tail"]

p_certain=len(certain_event)/len(coins)
print("certain_event(Head,Tail):",p_certain)
```
