## Multinomial Distribution

```python
import matplotlib.pyplot as plt
from math import factorial
from functools import reduce
import operator
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
n=10
count=[2,3,1,0,3,1] #1 to 6
p=[1/6]*6

#formula
denominator=reduce(operator.mul,(factorial(x) for x in count),1)
multinomial_coefficient=factorial(n)/denominator
print(multinomial_coefficient)

sequence_prob=reduce(operator.mul,(p[i]**count[i] for i in range(6)),1)
print(sequence_prob)

probability=multinomial_coefficient * sequence_prob
print(probability)
print(f"{1/probability:.0f}")

#plot
faces=[1,2,3,4,5,6]
plt.figure(figsize=(8,5))
sns.barplot(x=faces,y=count,palette="pastel")
plt.title("Dice roll outcomes count(10 rolls)")
plt.xlabel("Die Faces")
plt.ylabel("NO.of Occurrences")
plt.ylim(0,max(count)+1)
plt.text(2.5,max(count)+0.5,f"probability={probability:.6f}",fontsize=12,color="red")
plt.show
```

# Continuous  Distribution

## Continuous Probability Distribution

```python
from scipy.stats import norm
```

```python
# define parameters
#find the probability of students height b/w 160cm to 170cm

mu=165 # mean height in cm
sigma=10 # statndard devition
x=np.linspace(140,190) # continuous rnge of heights
print("Student Height:\n",x)
pdf=norm.pdf(x,mu,sigma)
print("Probability Density:\n",pdf)
#plot
plt.figure(figsize=(8,5))
plt.plot(x,pdf,color="blue",label="pdf")
plt.fill_between(x,pdf,0,where=(x>=160) &(x<=170),color="orange",alpha=0.5,
                 label="p(160 <= x <= 170)")
plt.title("Continous Probability Distribution :Student Heights")
plt.xlabel("Heights(cm)")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

## Uniform Distrubution

```python
from scipy.stats import uniform
```

```python
low=1 #(min dice value)
high=7

#generate 1000random numbers from uniform
dice_rolls_continuous=uniform.rvs(loc= low,scale=high-low,size=1000)
dice_rolls=np.floor(dice_rolls_continuous).astype(int)
print(dice_rolls)
frequencies=[np.sum(dice_rolls==i) for i in range(1,7)]
frequencies
for i,freq in enumerate(frequencies,start=1):
    print(f"Number {i}:{freq} times")

plt.bar(range(1,7),frequencies,color="lightgreen")
plt.xlabel("Dice Number")
plt.ylabel("Frequency")
plt.title("Dice Roll using scipy.stats.uniform")
plt.show()
```
