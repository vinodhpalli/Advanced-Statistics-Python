## join probability

```python
#rolling a dice
sample_space=[1,2,3,4,5,6]

A=[2,4,6] # even numbers
B=[4,5,6] #grter than 3

intersection=[i for i in sample_space if i in A and i in B ]
P_A_intersection_B=len(intersection)/len(sample_space)

print("Event A:",A)
print("Event B:",B)
print("A intersection B:",intersection)
print("P(A intersection B):",P_A_intersection_B)
```

## Contingeny tale/Cross tabulation table

```python
import pandas as pd

# create a contiguency tale as a dictornary
data={"Tea":[30,10],
     "coffee":[20,40]}
index=["male","female"]

# no of people who likes tea(male=30,female=10)
# no of people who likes coffe(male=20,female=40)

table=pd.DataFrame(data,index=index)
table["Total"]=table.sum(axis=1) #rowtotals
table.loc["Total"]=table.sum() #column totals
print(table)
grand_total=table.loc["Total","Total"]
print("--"*20)
prob_table=table/grand_total
print("probabilities Table:\n",prob_table)

#marginal probability
P_tea=prob_table.loc["Total","Tea"]
print("\nMarginal Probability:",P_tea)

#joint probability
p_male_tea=prob_table.loc["male","Tea"]
print("\njoint Probability:",p_male_tea)

#conditionl probability
p_male=prob_table.loc["male","Total"]
p_tea_given_male=p_male_tea/p_male
print("\nconditional Probability:",p_tea_given_male)



```
