### Non Negative Axiom

```python
sample_space={1,2,3,4,5,6}
probabilities={outcomes:1/6 for outcomes in sample_space}

for outcomes,prob in probabilities.items():
    print(f"p({outcomes})={prob} (>=0)")
print("--"*20)
```

### certain Axiom

```python
total_prob=sum(probabilities.values())
print("sum of probabillities:",total_prob)
```

### Aditivity Axiom

```python
sample_space={1,2,3,4,5,6}

#Each outome in dice has prob
prob=1/6

#event A:rolling an even number
A={2,4,6}

#event B:rolling an odd number
B={1,3,5}

#probabilities of A and b
p_A=len(A)*prob
p_B=len(B)*prob
p_union=len(A.union(B))*prob
p_union2=p_A + p_B

print(f"probabilitiy of A={p_A}")
print(f"probabilitiy of B={p_B}")
print(f"probabilitiy of AUB={p_union}")
print(f"probabilitiy of AUB={p_union2}")
```
