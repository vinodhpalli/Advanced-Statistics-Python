```python
import matplotlib.pyplot as plt
from math import factorial
import numpy as np
from scipy.stats import norm
import warnings
warnings.filterwarnings("ignore")
```

## Normal Distrubution

```python
scores=[55,60,62,65,68,70,70,72,75,78,80,82,85,88,90]

mu=np.mean(scores)
print("mean:",mu)
sigma=np.std(scores)
print("standard devition:",sigma)
```

```python
x=np.linspace(mu-3*sigma,mu+4*sigma,500)
pdf=norm.pdf(x,mu,sigma)
```

```python
plt.plot(x,pdf,color="black",lw=2)

plt.fill_between(x,pdf,0,where=(x>mu-sigma) & (x<mu+sigma),color="green",alpha=0.5,label="68% approx")
plt.fill_between(x,pdf,0,where=(x>mu-2*sigma) & (x<mu+2*sigma),color="yellow",alpha=0.3,label="95% approx")
plt.fill_between(x,pdf,0,where=(x>mu-3*sigma) & (x<mu+3*sigma),color="red",alpha=0.2,label="99.7% approx")

for i in range(-3,4):
    plt.axvline(mu+i*sigma,color="black",linestyle='--',lw=1)

plt.scatter(scores,[0]*len(scores),color="black",zorder=5,label="scores")
plt.axvline(mu,color="red",linestyle="--",label=f"mean={mu:.2f}")
plt.title("Normal Distrubition with Empirical Rule")
plt.xlabel("Scores")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```
