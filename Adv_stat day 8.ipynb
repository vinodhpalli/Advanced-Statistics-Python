# Sampling Techiniques

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "Employee":["A","B","C","D","E","F","G","H","I","J"],
    "Salary":[50,60,70,80,90,100,110,120,130,140]
}
df=pd.DataFrame(data)
print("Population Data:")
print(df)
```

```python
#Population mean
pop_mean=df["Salary"].mean()
print(pop_mean)
```

## 1).Convenience Sampling

```python
conv_sam=df.head(4)
print(conv_sam)
print("Sample Mean:",conv_sam["Salary"].mean())
```

## 2).Simple Random Sampling 

```python
simp_rand_sam=df.sample(n=4,random_state=1)
print(simp_rand_sam)

print(simp_rand_sam["Salary"].mean())
```

## 3). Systematic Sampling

```python
n=4
N=10
k=N/n
print(k)

#pick every seond employe starting from index 0
systematic_sample=df.iloc[::2][:4]
print(systematic_sample)
print("systematic_sample mean:",systematic_sample["Salary"].mean())
```

## 4). stratifed sampling


```python
low=df[df["Salary"]<80].sample(n=1,random_state=3)
print(low)
medium=df[(df["Salary"]>=80) & (df["Salary"]<=120)].sample(n=2,random_state=2)
print(medium)
high=df[df["Salary"]>120].sample(n=1,random_state=3)
print(high)

strat_samp=pd.concat([low,medium,high])
print(strat_samp)

print("sample mean:",strat_samp["Salary"].mean())
```

## 5).Cluster Sample

```python
# definre cluster

cluster1=df.iloc[0:3]
print(cluster1)
cluster2=df.iloc[3:7]
print(cluster2)
cluster3=df.iloc[7:11]
print(cluster3)

clusters=[cluster1,cluster2,cluster3]
np.random.seed(1)
selected_cluster=clusters[np.random.randint(0,len(clusters))]
print(selected_cluster)

print("Sample Mean:",selected_cluster["Salary"].mean())
```

# Sampling Distrubution

```python
data={
    "Employee":["A","B","C","D","E","F","G","H","I","J"],
    "Salary":[50,60,70,80,90,100,110,120,130,140]
}
df=pd.DataFrame(data)
print("Population Data:")
print(df)
```

```python
#Population mean
pop_mean=df["Salary"].mean()
print(pop_mean)
```

```python
np.random.seed(1)

sample_size=2
num_samples=10000
sample_means=[]
for i in range(num_samples):
    sample=df["Salary"].sample(n=sample_size,replace=True)
    sample_means.append(sample.mean())
sample_means=np.array(sample_means)
mean_samp_mean=sample.mean()
plt.figure(figsize=(10,5))
plt.hist(sample_means,bins=10,color="skyblue",edgecolor="black")
plt.axvline(pop_mean,color='red',linestyle='dashed',linewidth=2,label=f"popultion mean={pop_mean}")
plt.axvline(mean_samp_mean,color='blue',linestyle='dashed',linewidth=2,label=f"mean of sample means={mean_samp_mean}")
plt.title("sampling Distrubution of sample mean (sample size=2)")
plt.xlabel("Sample mean Salary")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
