## Discrete Probability Distrubution (PMF,CDF)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
scores=[1,2,2,3,3,3,4,4,5]

unique_scores=sorted(set(scores))
print("Unique scores:",unique_scores)

#caluclate PMF (probability of each unique scores)
pmf=[scores.count(x)/len(scores) for x in unique_scores]

#Caluclate CDF(cumultive sum of PMF values)
cdf=np.cumsum(pmf)
print()
print("Score | pmf | cdf")
for  x,p,f in zip(unique_scores,pmf,cdf):
    print(f"{x}      | {p:3f} |{f:3f}")

# Plot pmf using bar chat
plt.bar(unique_scores,pmf,color="skyblue",edgecolor="black",label="PMF")

#plot cdf using step
plt.step(unique_scores,cdf,where="post",color="red",linewidth=2,label="CDF")

plt.title("Discrete Probability Distrubution (PMF,CDF)")
plt.xlabel("Score")
plt.ylabel("Probability")
plt.ylim(0,1.05)
plt.legend()
plt.grid(True)
plt.show()
```

## Binomial Random Varible & Distribution

```python
import math
```

```python
n=5
p=0.7
x_values=list(range(n+1)) #possible no.of sucess

def binomial_prob(n,k,p):
    comb=math.factorial(n)/(math.factorial(k)*math.factorial(n-k))
    prob=comb*(p**k)*((1-p)**(n-k))
    return prob
probabilities=[]
for k in x_values:
    prob=binomial_prob(n,k,p)
    probabilities.append(prob)
    print(f"P(X={k})={prob:.4f}")

plt.bar(x_values,probabilities,color="skyblue",edgecolor="black")
plt.title("Binomial Random Varible & Distribution")
plt.xlabel("Number of exams passed (x)")
plt.ylabel("probability")
plt.xticks(x_values)
plt.ylim(0,max(probabilities)+0.05)
plt.show()

```
